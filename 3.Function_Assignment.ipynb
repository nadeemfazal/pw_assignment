1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
  sum_of_evens = 0
  for number in numbers:
    if isinstance(number, int) and number % 2 == 0:
      sum_of_evens += number
  return sum_of_evens
```

```python
number = [1, 4, 5, 7,8,10]
sum_even_numbers(number)
```

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
def square_numbers(numbers):
  l = []
  for i in numbers:
    l.append(i**2)
  return l

li = [1, 2, 3, 4, 5]
squared_list = square_numbers(li)
print(squared_list)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(number)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next_value = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            value = self.current
            self.current, self.next_value = self.next_value, self.current + self.next_value
            self.count += 1
            return value
        else:
            raise StopIteration

fib = FibonacciIterator(10)
for number in fib:
    print(number)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent


```python
def powers_of_two(x):

    for i in range(x + 1):
        yield 2 ** i

for power in powers_of_two(10):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
  """
  Reads a file line by line and yields each line as a string.

  Args:
    file_path: The path to the file to read.

  Yields:
    Each line in the file as a string.
  """
  try:
    with open(file_path, 'r') as file:
      for line in file:
        yield line.strip()  # Remove leading/trailing whitespace
  except FileNotFoundError:
    print(f"Error: File not found at '{file_path}'")

# Example usage:
# Assuming you have a file named 'example.txt' in your current directory
# for line in read_file_line_by_line('example.txt'):
#   print(line)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.


```python

def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'

    filtered_chars = filter(lambda char: char not in vowels, input_string)

    return ''.join(filtered_chars)


input_string = "Hello, welcome to pwskills"
result = remove_vowels(input_string)
print(result)
```

product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.

```python

orders = [
    [34587, 40.95, 4],  # Order 1: learning pyhton, mark lutz
    [98762, 56.80, 5],  # Order 2: programing pyhton, Mark Lutz
    [98762, 32.95, 3],  # Order 3: Head First pyhton, Paul Berry
    [88112, 24.99, 3]   # Order 4: einfuhrung in pyhton3, Bernd Klein
]

# Function to calculate total price for each order
def calculate_order(order):
    order_number, price_per_item, quantity = order
    total_price = price_per_item * quantity
    # Add €10 if the total price is less than €100
    if total_price < 100:
        total_price += 10
    return (order_number, total_price)

# Using map to process the orders
result = list(map(calculate_order, orders))

# Example usage
print(result)
```
